```python
no_sleep_disorder=df['Sleep Disorder'].value_counts().to_frame().reset_index()
no_sleep_disorder
```

```python
new=df.groupby(['Age','Gender','Occupation','BMI Category','Daily Steps']).agg({'Sleep Duration':'mean','Stress Level':'mean','Heart Rate':'mean','Quality of Sleep':'mean','Physical Activity Level':'mean'}).reset_index()
```

```python
new
```

```python
new['Age Category']=pd.cut(new['Age'],bins=[20,30,40,50,60],labels=['junior','senior','super senior','retirement'])
new
```

```python
age=new['Age Category'].value_counts().to_frame().reset_index()
age
```

### SLEEP DURATION OVER AGE

```python

plt.figure(figsize=(15,8))
plt.bar(new['Age'],new['Sleep Duration'],color='black')
plt.title('Sleep duration over age',color='red')
plt.xlabel('Age')
plt.ylabel('Avg sleep duration')

plt.show()
```

### STRESS LEVEL OVER TIME

```python
plt.figure(figsize=(15,8))
plt.bar(new['Age'],new['Stress Level'],color='grey')
plt.title('Stress level over time',color='red')
plt.xlabel('Age')
plt.ylabel('Avg stress level')
plt.show()
```

```python
### QUALITY OF SLEEP OVER AGE
```

```python
plt.figure(figsize=(15,8))
plt.bar(new['Age'],new['Quality of Sleep'],color='skyblue')
plt.title('QUALITY OF SLEEP OVER AGE',color='blue')
plt.xlabel('Age')
plt.ylabel('Avg quality of sleep')
plt.show()
```

### AVG SLEEP DURATION BY OCCUPATION

```python
plt.figure(figsize=(15,8))
plt.bar(new['Occupation'],new['Sleep Duration'],color='pink')
plt.title('AVG SLEEP DURATION BY OCCUPATION',color='red')
plt.xlabel('Occupation')
plt.ylabel('Avg sleep duration')
plt.xticks(rotation=90)
plt.show()
```

### PHYSICAL ACTIVITY LEVEL BY GENDER

```python
plt.figure(figsize=(4,4))
plt.bar(new['Gender'],new['Physical Activity Level'],color='yellow')
plt.title('PHYSICAL ACTIVITY LEVEL BY GENDER',color='green')
plt.xlabel('Gender')
plt.ylabel('Avg Physical Activity Level')
plt.show()
```

### STRESS LEVEL BY BMI CATEGORY

```python
plt.figure(figsize=(8,6))
plt.bar(new['BMI Category'],new['Stress Level'],color='grey')
plt.title('STRESS LEVEL BY BMI CATEGORY',color='blue')
plt.xlabel('BMI Category')
plt.ylabel('Avg Stress Level')
plt.xticks(rotation=90)
plt.show()
```

### HEART RATE OVER DAILY STEPS

```python
plt.figure(figsize=(10,8))
plt.bar(new['Heart Rate'],new['Daily Steps'],color='green')
plt.title('HEART RATE OVER DAILY STEPS')
plt.xlabel('Avg Heart Rate')
plt.ylabel('Avg Daily Steps')
plt.show()
```

## FREAQUENCY OF SLEEP DISORDERS BY AGE GROUP

```python
plt.figure(figsize=(10,8))
plt.plot(age['Age Category'],age['count'],color='green',marker='*')
plt.title('FREAQUENCY OF SLEEP DISORDERS BY AGE GROUP')
plt.xlabel('Category')
plt.ylabel('Age')
plt.show()
```

### Sleep Duration Distribution

```python
plt.figure(figsize=(12,6))
plt.title('Sleep Duration Distribution')
plt.hist(new['Sleep Duration'],bins=10,color='black',edgecolor='red')
plt.xlabel('Avg Sleep Duration')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Heart Rate Distribution

```python
plt.figure(figsize=(10,6))
plt.title('Heart Rate Distribution')
plt.hist(new['Heart Rate'],bins=10,color='grey',edgecolor='black')
plt.xlabel('Avg Heart Rate')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Stress Level Distribution

```python
plt.figure(figsize=(10,6))
plt.title('Stress Level Distribution',color='red')
plt.hist(new['Stress Level'],bins=10,color='orange',edgecolor='black')
plt.xlabel('Avg Stress Level')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
